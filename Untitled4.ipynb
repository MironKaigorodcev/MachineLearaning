```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
import os

train_dir = '/content/classification-of-butterflies/train_butterflies/train_split'

# Преобразования для тренировочных данных
train_transforms = transforms.Compose([
    #transforms.RandomResizedCrop(224),  # Обрезка и изменение размера
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.RandomPerspective(distortion_scale=0.3),# Случайное отражение по горизонтали
    transforms.ToTensor(),  # Преобразование в тензор
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Нормализация
])

# Загрузка данных
train_dataset = datasets.ImageFolder(train_dir, transform=train_transforms)
key = train_dataset.class_to_idx
train_dataset1, val_dataset = train_test_split(train_dataset, test_size=0.05, shuffle=True)

# Создание DataLoader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader= DataLoader(val_dataset, batch_size=32, shuffle=True, num_workers=2)

# Проверка данных
for images, labels in train_loader:
    print(images)  # Размер батча: (batch_size, channels, height, width)
    print(labels)
      # Метки классов
    break
print(key)
```

```python
def train_model(model, loss, optimizer, scheduler, num_epochs):

    loss_hist = {'train':[], 'val':[]}
    acc_hist = {'train':[], 'val':[]}

    best_val_loss = float('inf')

    for epoch in range(num_epochs):
        print("Epoch {}/{}:".format(epoch, num_epochs - 1), end="")
        for phase in ['train', 'val']:
            if phase == 'train':
                dataloader = train_loader
                scheduler.step()
                model.train()
            else:
                dataloader = val_loader

            running_loss = 0.
            running_acc = 0.

            # Итерируемся по dataloader
            for inputs, labels in tqdm(dataloader):
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    preds = model(inputs)
                    loss_value = loss(preds, labels)
                    preds_class = preds.argmax(dim=1)

                    if loss_value < best_val_loss:
                      best_val_loss = loss_value
                      best_model_weights = model.state_dict()

                    if phase == 'train':
                        loss_value.backward() # Считаем градиент
                        optimizer.step() # Считаем шаг градиентного спуска

                # Статистика
                running_loss += loss_value.item() #считаем Loss
                running_acc += (preds_class == labels.data).float().mean().data.cpu().numpy()  #считаем accuracy

            epoch_loss = running_loss / len(dataloader)  # Loss'ы делим на кол-во бачей в эпохе
            epoch_acc = running_acc / len(dataloader) #считаем Loss на кол-во бачей в эпохе

            print("{} Loss: {:.4f} Acc: {:.4f} ".format(phase, epoch_loss, epoch_acc), end="")

            loss_hist[phase].append(epoch_loss)
            acc_hist[phase].append(epoch_acc)

    return model, loss_hist, acc_hist
```

```python
#model = models.resnet152(pretrained=True)
model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)

# Замена классификатора
num_classes = 50  # Указываем нужное количество классов
in_features = model.classifier[1].in_features  # Получаем количество входных признаков
model.classifier[1] = torch.nn.Linear(in_features, num_classes)  # Заменяем последний слой
for param in model.parameters():
  param.requires_grad = True

#model.fc = torch.nn.Linear(model.fc.in_features, 50)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)

loss = torch.nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), amsgrad=True, lr=0.001)

scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
```

```python
#за 8 эпох вполне обучается до адекватного уровня
model, loss, acc = train_model(model, loss, optimizer, scheduler, num_epochs=10)
```

```python
plt.rcParams['figure.figsize'] = (14, 7)
for experiment_id in acc.keys():
    plt.plot(acc[experiment_id], label=experiment_id)
plt.legend(loc='upper left')
plt.title('Model Accuracy')
plt.xlabel('Epoch num', fontsize=15)
plt.ylabel('Accuracy value', fontsize=15);
plt.grid(linestyle='--', linewidth=0.5, color='.7')
```

```python
test_dir = '/content/classification-of-butterflies/test_butterflies'

class ImageFolderWithPaths(torchvision.datasets.ImageFolder):
    def __getitem__(self, index):
        original_tuple = super(ImageFolderWithPaths, self).__getitem__(index) #Дополняем original_tuple путем для файла (.__getitem__(index))
        path = self.imgs[index][0]
        tuple_with_path = (original_tuple + (path,))
        return tuple_with_path

# Преобразования для тренировочных данных
test_transforms = transforms.Compose([
    transforms.ToTensor(),  # Преобразование в тензор
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Нормализация
])


test_dataset = ImageFolderWithPaths(test_dir, test_transforms) #Берем новый класс и получаем tuple из 3х значений
test_loader = torch.utils.data.DataLoader(
        test_dataset, batch_size=32, shuffle=False, num_workers=0)

# Проверка данных
for images, labels, path in test_loader:
    print(images.shape)  # Размер батча: (batch_size, channels, height, width)
    print(labels)
    print(path)# Метки классов
    break
```

```python
model.eval()

test_predictions = []  #Создаем пустой список предсказания
test_img_paths = [] #Пути до изображения

for inputs, labels, paths in tqdm(test_loader): #Цикл по test_dataloader inputs - батч с изображением, lable - тут none, paths - пути до изображения
  inputs = inputs.to(device)
  labels = labels.to(device)
  with torch.set_grad_enabled(False):
    preds = model(inputs) # Считаем предикшены
    test_predictions.extend(preds.argmax(dim=1).cpu().numpy())
    test_img_paths.extend(paths)

test_predictions
```

```python
submission_df = pd.DataFrame.from_dict({'index': test_img_paths, 'label': test_predictions})
submission_df['index'] = submission_df['index'].str.replace('/content/classification-of-butterflies/test_butterflies/valid/', '')
submission_df['index'] = submission_df['index'].str.replace('.jpg', '')
submission_df.set_index('index', inplace=True)
submission_df.astype(int)
```
